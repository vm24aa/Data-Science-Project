```python
# EDA
```

```python
# Detect target column automatically
target_col = [col for col in df.columns if col.lower() in ['status','class','label','target']]
if target_col:
    target = target_col[0]
    print(f" Target Column Detected: {target}")
else:
    raise ValueError("Target column not found. Please specify manually.")
```

```python
#  Class Balance
plt.figure(figsize=(5,4))
sns.countplot(x=target, data=df, palette='coolwarm')
plt.title("Class Distribution (Parkinson’s vs Healthy)")
plt.xlabel("Class (0 = Healthy, 1 = Parkinson’s)")
plt.ylabel("Count")
plt.show()
```

```python
# Males and Females in each class

sns.heatmap(pd.crosstab(df['class'], df['gender']).divide(3).astype('int64'),
            yticklabels=['No PD', 'PD'],
            xticklabels=['Female', 'Male'],
            annot=True,
            fmt='d')
plt.title('Number of males and females in each class')
plt.show()
```

```python
#  Correlation with Target (Bar Chart)
corr = df.corr(numeric_only=True)
plt.figure(figsize=(8,5))
corr_target = corr[target].sort_values(ascending=False)[1:21]
sns.barplot(x=corr_target.values, y=corr_target.index, palette="viridis")
plt.title("Top 20 Features Correlated with Target")
plt.xlabel("Correlation with Parkinson’s Status")
plt.ylabel("Features")
plt.show()
```
